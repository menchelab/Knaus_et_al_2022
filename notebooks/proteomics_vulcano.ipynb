```python
import pandas as pd
proteomics = pd.read_excel(
    '../raw/Proteomics_vulcano_F-test_input_DM.xlsx',
)
proteomics.rename(
    columns = {
        'F-test: Mean log2(Ratio) - KO_grp1 vs WT': 'log2FC',
        'F-test: Moderated -log10(Pvalue)': '-log10(pvalue)'
    },
    inplace = True
)
proteomics
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
fig, ax = plt.subplots()
pval_threshold = 0.01
data = proteomics
lfc_threshold_up = 0.43
lfc_threshold_down = -0.42
print(lfc_threshold_down, lfc_threshold_up)
hue_idx_up = (data['log2FC'].astype(float) >= lfc_threshold_up) & \
    (data['-log10(pvalue)'].astype(float) > -np.log10(pval_threshold))
hue_idx_down = (data['log2FC'].astype(float) <= lfc_threshold_down) & \
    (data['-log10(pvalue)'].astype(float) > -np.log10(pval_threshold))

#hue_map = [s.metabolitegroupcoarse if sig else 'none' for (i, s), sig in zip(data.iterrows(), hue_idx)]
# palette_dict = {
#     k: sns.color_palette('husl', len(np.unique(hue_map)))[i] for i, k in enumerate(np.unique(hue_map))
# }
#palette_dict['none'] = 'lightgrey'
hue = []
for up, down in zip(hue_idx_up, hue_idx_down):
    if up and not down:
        hue.append('up')
    
    elif not up and down:
        hue.append('down')
    
    else:
        hue.append('none')
    
sns.scatterplot(
    x = data['log2FC'].astype(float),
    y = data['-log10(pvalue)'].astype(float),
    ax = ax,
    hue = hue,
    palette = {
        'down': '#2E2D7E',
        'up': '#BE1824',
        'none': 'lightgrey'
    },
    s = 50
)

for x in [
    lfc_threshold_down, 
    lfc_threshold_up
]:
    ax.axvline(
        x = x,
        ls = '--',
        lw = 0.5,
        c = 'k'
    )

ax.set_xlabel('log2FC')
ax.set_ylabel('-log10(adj. pvalue)')

ax.axhline(
    y = -np.log10(pval_threshold),
    ls = '--',
    lw = 0.5,
    c = 'k'
)
ax.set_xlim(-9.5, 9.5)

fig.set_figwidth(5)
fig.set_figheight(5)
fig.tight_layout()
fig.savefig('../plots/proteomics_vulcano_50.pdf')
```
